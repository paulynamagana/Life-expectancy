## Import libraries

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import InputLayer, Dense
from tensorflow.keras.optimizers import Adam
```

## Import dataset

```python
dataset = pd.read_csv("life_expectancy.csv", )
dataset.head() #inspect the data
```

```python
#get stas from the data
dataset.describe()
```

Year, we can see the data goes from 2000 to 2015

```python
dataset.drop(["Country"], axis=1, inplace=True)
```

```python
labels = dataset.iloc[:,-1] #extract labels, life expectancy
features = dataset.iloc[:,0:-1]
```

```python
features = pd.get_dummies(features)
features.head()
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.20, random_state=23)
```

```python
numerical_features = features.select_dtypes(include=["float64", "int64"])
numerical_columns = numerical_features.columns
ct = ColumnTransformer([("scaler", StandardScaler(), numerical_columns)], remainder = "passthrough")
```

```python
features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.transform(features_test)
```

```python
my_model = Sequential()
```

```python
num_features = features.shape[1]
input = InputLayer(input_shape = (num_features,))
my_model.add(input)
my_model.add(Dense(64, activation="relu"))
my_model.add(Dense(1))
my_model.summary()
```

```python
opt = Adam(learning_rate= 0.01)
my_model.compile(loss="mse", metrics = ["mae"], optimizer=opt)
```

```python
my_model.fit(features_train_scaled, labels_train, epochs = 40, batch_size= 1, verbose=1)
```

```python
res_mse, res_mae = my_model.evaluate(features_test_scaled, labels_test, verbose=0)
```

```python
print("MAE: ", res_mae)
print("RMSE: ", res_mse)
```
